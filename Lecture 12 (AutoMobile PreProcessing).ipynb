```python
dataset[dataset['make'] == 'alfa-romero'].loc[:,'normalized-losses']
```

```python
dataset['normalized-losses'].isnull()
```

```python
na_ind = []
for index_row in range(len(dataset)):
    if str(dataset.loc[index_row]['normalized-losses'])== 'nan':
        na_ind.append(index_row)
```

```python
na_ind
```

```python
normalized_mis_values=dataset.iloc[na_ind,:]
```

```python
normalized_mis_values
```

```python
brands = normalized_mis_values['make'].unique()
```

```python
brands_dic = {}
for brand in brands:
    values = dataset[dataset['make']==brand].loc[:,'normalized-losses']
    brands_dic[brand]=values.median()
```

```python
brands_dic
```

```python
dataset.isna().sum()
```

```python
for index_row in range(len(dataset)):
    if dataset.loc[index_row,'make'] in brands_dic.keys():
        if str(dataset.loc[index_row,'normalized-losses']) == 'nan':
            dataset.loc[index_row,'normalized-losses'] = brands_dic[dataset.loc[index_row, 'make']]
```

```python
dataset.isna().sum()
```

```python
dataset['normalized-losses'].fillna(pd.to_numeric(dataset['normalized-losses']).mean(),inplace=True)
```

```python
dataset.isna().sum()
```

```python
dataset[(dataset['bore']).isna()]
```

```python
dataset[pd.to_numeric(dataset['bore']).isna()].index
```

```python
mazda_bore = dataset[dataset['make']=='mazda']
```

```python
mazda_median_bore = mazda_bore.loc[:,'bore'].median()
```

```python
for row_index in range(len(dataset)):
    if dataset.loc[row_index, 'make'] == 'mazda':
        if dataset.loc[row_index, 'bore'] == 'nan':
            dataset.loc[row_index, 'bore'] = mazda_median_bore
```

```python
dataset['bore'].fillna(mazda_median_bore,inplace=True)
```

```python
dataset[(dataset['stroke']).isna()]
```

```python
dataset[pd.to_numeric(dataset['stroke']).isna()].index
```

```python
mazda_stroke = dataset[dataset['make'] == 'mazda']
```

```python
mazda_median_stroke = mazda_stroke.loc[:, 'bore'].median()
```

```python
for row_index in range(len(dataset)):
    if dataset.loc[row_index, 'make'] == 'mazda':
        if dataset.loc[row_index, 'bore'] == 'nan':
            dataset.loc[row_index, 'bore'] = mazda_median_stroke
```

```python
dataset['stroke'].fillna(mazda_median_stroke, inplace=True)
```
