```python

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
x = np.arange(0,3*np.pi,0.1)
y = np.sin(x)
plt.plot(x,y)
plt.title("Pi - Sin(Pi)")
plt.xlabel("X")
plt.ylabel("Y")
plt.style.use('seaborn-white')
plt.show()
```

```python
x = np.arange(0,3*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(1,2,1)
plt.title("Sin")
plt.plot(x,y_sin,'r', linewidth=5, linestyle='dotted')

plt.subplot(1,2,2)
plt.title("Cos")
plt.plot(x,y_cos,'b', linewidth=5, linestyle='-.')

plt.figure(figsize=(10,10))
plt.show()
```

```python
fig, axs = plt.subplots(1,3,figsize = (20,8))
x = np.linspace(0,10,500)
y_sin = np.sin(x)
y_cos = np.cos(x)
y_tan = np.tan(x)

axs[0].plot(x,y_sin,'r')
axs[0].set_title("Sin Figure")
axs[0].set_xlabel("X")
axs[0].set_ylabel("Sin")

axs[1].plot(x,y_cos,'b')
axs[1].set_title("Cos Figure")
axs[1].set_xlabel("X")
axs[1].set_ylabel("Cos")

axs[2].plot(x,y_tan,'g')
axs[2].set_title("Tan Figure")
axs[2].set_xlabel("X")
axs[2].set_ylabel("Tan")

plt.figure(figsize=(3,2))
plt.show()
```
