```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dataset.FUELTYPE = le.fit_transform(dataset.FUELTYPE)
dataset.Brands = le.fit_transform(dataset.Brands)
dataset.VEHICLECLASS = le.fit_transform(dataset.VEHICLECLASS)
```

```python
target = dataset.CO2EMISSIONS
features = dataset.iloc[:, :-1]
```

```python
features
```

```python
dataset.drop(['CO2EMISSIONS'], axis=1, inplace=True)
```

```python
plt.scatter(features.Brands,target)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features.values, target.values, test_size=0.1)
```

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error

r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
```

```python
mse
```

```python
r2
```

```python
from sklearn.svm import SVR
svr = SVR(kernel='rbf')
svr.fit(X_train,y_train)

from sklearn.tree import DecisionTreeRegressor
dtreg = DecisionTreeRegressor()
dtreg.fit(X_train,y_train)
r2_score(y_test,dtreg.predict(X_test)) ,r2_score(y_train,dtreg.predict(X_train))
```

```python
r2_score(y_test,svr.predict(X_test))
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 2)
X_poly = poly_reg.fit_transform(features.values)
X_train, X_test, y_train, y_test = train_test_split(X_poly, target.values, test_size=0.1)
pol = LinearRegression()
pol.fit(X_train, y_train)
```

```python
poly_pred = pol.predict(X_test)
train_pred = pol.predict(X_train)
```

```python
r2_score(y_test, poly_pred)*100
```

```python
r2_score(y_train,train_pred)*100
```

```python
train_data.drop(['PoolQC','Fence','MiscFeature','Alley'],axis=1,inplace=True)
```

```python
from sklearn.impute import SimpleImputer

for col in train_data.columns:
    if train_data[col].dtype == 'object':
        s = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        train_data[col]=s.fit_transform(train_data.loc[:,col].values.reshape(-1,1))
    else:
        s = SimpleImputer(missing_values=np.nan, strategy='mean')
        train_data[col]=s.fit_transform(train_data.loc[:,col].values.reshape(-1,1))
```
