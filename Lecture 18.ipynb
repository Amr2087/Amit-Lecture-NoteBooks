### Handling Outliers

```python
def z_score(data,col):
    mean = data[col].mean()
    std= data[col].std()
    outliers = []
    for i in range (len(data)):
        if data.loc[i,col]>mean+3*std or data.loc[i,col]<mean-3*std:
            outliers.append(data.loc[i,col])
    print(f'Outlieres in Column {col} : {outliers}')
```

```python
columns = list(data.columns)
columns.remove('Species')
columns.remove('Weight')
columns
```

```python
for col in columns:
    z_score(data,col)
```

```python
def handle(data,col):
    mean = data[col].mean()
    std= data[col].std()
    high = mean+3*std
    low = mean-3*std
    for i in range (len(data)):
        if data.loc[i,col]>high:
            data.loc[i,col] = high
        elif data.loc[i,col]<low:
            data.loc[i,col] = low
```

```python
for col in columns:
    handle(data,col)
```

```python
data.drop([40],axis=0,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data.Species = le.fit_transform(data.Species)
```

```python
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.1,random_state=1)
```

### Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(n_estimators = 20, max_depth =5,min_samples_split=3)
rfr.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score
y_pred = rfr.predict(X_test)
r2 = r2_score(y_test,y_pred)
print(f'Score : {r2*100}%')
```

```python
y_pred = rfr.predict(X_train)
r2 = r2_score(y_train,y_pred)
print(f'Score : {r2*100}%')
```

### Linear regression

```python
from sklearn.linear_model import LinearRegression
l_reg = LinearRegression()
l_reg.fit(X_train,y_train)
```

```python
y_preds  = l_reg.predict(X_test)
r22 = r2_score(y_test,y_preds)
print(f'Score : {r22*100}%')
```

```python
y_preds  = l_reg.predict(X_train)
r22 = r2_score(y_train,y_preds)
print(f'Score : {r22*100}%')
```

### Grid search

```python
grid_params = {'max_depth' : [2,3,4],
               'n_estimators' : [20,50,100],
               'min_samples_split' :[2,3,4,5]}
```

```python
from sklearn.model_selection import GridSearchCV
rfr = RandomForestRegressor()
gridsearch = GridSearchCV(estimator=rfr,param_grid=grid_params,cv=5,n_jobs=-1)
gridsearch.fit(X_train, y_train)
```

```python
gridsearch.best_params_
```

```python
ypred = gridsearch.predict(X_test)
```

```python
r222 = r2_score(y_test,ypred)
```

```python
r222
```

```python
model = RandomForestRegressor(**gridsearch.best_params_)
model.fit(X_train, y_train)
```

```python
new_pred = model.predict(X_test)
```

```python
r = r2_score(y_test,new_pred)
```

```python
r
```

### Cross validation

```python
from sklearn.model_selection import KFold,cross_val_score
model = LinearRegression()
cv = KFold(n_splits=10, shuffle=True)
scores = cross_val_score(model,X_train, y_train,cv=cv,scoring='neg_mean_squared_error',n_jobs=-1)

print("Cross validation accuracy  : ",np.mean(scores))
```
