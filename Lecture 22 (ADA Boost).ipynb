# Filter Methods

### Linear Correlation

```python
plt.figure(figsize=(12,9))
sns.heatmap(dataset.corr(), annot=True)
```

```python
corr_list = dataset.corr().target.apply(lambda  x:abs(x) > 0.1)
# First Method
columns = [v for v in corr_list.index if corr_list[v]]
columns.pop()
```

```python
columns
```

```python
# Second Method
ind = []
for i in range(len(corr_list)):
    if corr_list[i]:
        ind.append(i)
```

```python
ind.pop()
```

```python
ind
```

```python
X = dataset.loc[:, columns]
```

### Variance Threshold

# Wrapper Methods

## Forward Elimination

```python
X_main
```

```python
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import  LogisticRegression
import warnings
warnings.filterwarnings('ignore')
clf  = LogisticRegression()
sfs = SequentialFeatureSelector(estimator=clf,direction='forward',scoring='accuracy')
sfs.fit(X_main,y)
```

```python
sfs.get_feature_names_out().tolist()
```

```python
sfs = SequentialFeatureSelector(estimator=clf,direction='backward',scoring='accuracy')
sfs.fit(X_main,y)
sfs.get_feature_names_out().tolist()
```

```python
from mlxtend.feature_selection import ExhaustiveFeatureSelector
efs = ExhaustiveFeatureSelector(scoring='accuracy',estimator=clf,cv = 5, max_features=10, min_features=5,n_jobs=-1)
efs.fit(X_main,y)
efs.best_feature_names_
```

## Embedded Method

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100)
model.fit(X_main,y)

importance = model.feature_importances_
```

```python
importance
```

# PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=8)
components = pca.fit_transform(X_main)
```

```python
pca.explained_variance_ratio_.sum()
```

```python
n_data = pd.DataFrame(data=components)
```

```python
n_data['target'] = y
```
