```python
import matplotlib.pyplot as plt
import seaborn as sea
import numpy as np
```

```python
x = np.arange(0,3*np.pi,0.1)
y = np.sin(x)
plt.plot(x,y)
plt.title("Pi - Sin(Pi)")
plt.xlabel("X")
plt.ylabel("Y")
plt.style.use('seaborn-white')
plt.show()
```

```python
x = np.arange(0,3*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(1,2,1)
plt.title("Sin")
plt.plot(x,y_sin,'r', linewidth=5, linestyle='dotted')

plt.subplot(1,2,2)
plt.title("Cos")
plt.plot(x,y_cos,'b', linewidth=5, linestyle='-.')

plt.figure(2)
plt.show()
```
