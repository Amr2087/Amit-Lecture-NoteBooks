```python
def z_score(data, col):
    mean = data[col].mean()
    std = data[col].std()
    outliers = []
    for i in range(len(data)):
        if data.loc[i, col] > mean + 3 * std or data.loc[i, col] < mean - 3 * std:
            outliers.append(data.loc[i, col])
    print(f'Outlieres in Column {col} : {outliers}')
```

```python
columns = list(X.columns)
columns
```

```python
for col in columns:
    z_score(dataset, col)
```

```python
def handle(data, col):
    mean = data[col].mean()
    std = data[col].std()
    high = mean + 3 * std
    low = mean - 3 * std
    for i in range(len(data)):
        if data.loc[i, col] > high:
            data.loc[i, col] = high
        elif data.loc[i, col] < low:
            data.loc[i, col] = low
```

```python
for col in columns:
    handle(dataset, col)
```

```python
import pandas as pd
data = pd.read_csv('data/Car Fuel Consumption.csv.xls')
```

```python
data.drop(['refill liters', 'refill gas', 'specials'],inplace=True,axis=1)
```

```python
data['consume']=data['consume'].str.replace(',', '.').astype(float)
data['distance']=data['distance'].str.replace(',', '.').astype(float)
data['temp_inside']=data['temp_inside'].str.replace(',', '.').astype(float)
```

```python
for i in data[data['temp_inside'].isna()].index:
    data.loc[i,'temp_inside'] = data['temp_inside'].mean()
```

```python
data.info()
```

```python
y = data.consume.values
X = data[['distance','speed','temp_inside','temp_outside','gas_type','AC','rain','sun']].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(X_train,y_train)
```

```python
ye_pred = regression.predict(X_test)
yr_pred = regression.predict(X_train)
```

```python
from sklearn.metrics import r2_score
r2_ye = r2_score(y_test,ye_pred)
r2_yr = r2_score(y_train,yr_pred)
```
